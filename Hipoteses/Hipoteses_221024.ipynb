## Análises gerais

```python
dados = pd.read_csv('salarios_profs.csv', encoding = 'latin1', sep= ';', decimal=',')
dados.head()
```

```python
unidades = ['ICMC', 'FEA', 'IME']
jornada = 'RDIDP'
funcao = 'Professor Titular'
dados_filtrados = dados[(dados['UNID_ORGAO'].isin(unidades)) & \
                        (dados['JORNADA'] == jornada) & \
                        (dados['FUNCAO'] == funcao)]
dados_filtrados.head()
```

```python
dados_filtrados.groupby('UNID_ORGAO')['SALARIO_MENSAL_LIQUIDO'].describe()
```

#Testes

- Teste 1

Ho: Salario IME <= 20k

Ha: Salario IME > 20k

```python
unidade = 'IME'
salario_hipotese = 20000
salario_unidade = dados_filtrados[dados_filtrados['UNID_ORGAO'] == unidade]['SALARIO_MENSAL_LIQUIDO']
```

```python
st.ttest_1samp(a = salario_unidade, popmean = salario_hipotese, alternative = 'greater')
```

- Teste 2

H0: S-ime = S-icmc

Ha: S-ime $\neq$ S-icmc

```python
grupo_a = 'IME'
grupo_b = 'ICMC'
df_a = dados_filtrados[dados_filtrados['UNID_ORGAO'] == grupo_a]['SALARIO_MENSAL_LIQUIDO']
df_b = dados_filtrados[dados_filtrados['UNID_ORGAO'] == grupo_b]['SALARIO_MENSAL_LIQUIDO']
```

```python
st.ttest_ind(a = df_a, b = df_b, alternative = 'two-sided')
```
